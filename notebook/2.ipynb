# Inputs

## QST 1
Using input function, create a list base on the input provided by user and get inputs for the list too

````python
n = int(input("enter list length: "))
arr = []
for i in range(n):
    val=int(input("enter input for index " + str(i) + " : "))
    arr.append(val)

print(arr)
````

using while loop

````python
n = int(input("enter list length: "))
arr = []
while n>0:
    val = int(input("enter input for index " + str(n) + " : "))
    arr.append(val)
    n=n-1
print(arr)
````

using list comprehension

````python
n,arr = int(input("enter list length: ")),[]
[arr.append(int(input("enter input for index " + str(i) + " : "))) for i in range(n)]
print(arr)
````

**Efficient way**:
Often in competitive programming or data processing, it's more efficient to ask the user for all numbers at once (separated by spaces) rather than hitting the input() syscall n times.

````python
# Input: 10 20 30 40
arr = list(map(int, input("Enter all numbers separated by space: ").split()))
````

Usage of while

````python
num = int(input("Enter the num, (0 to stop)"))
while num != 0:
    print("Entered number is: ", num)
    num=int(input("Enter the num, (0 to stop)"))
````

````python
num = int(input("Enter the num, (0 to stop)"))
while True:
    print("Entered number is: ", num)
    num=int(input("Enter the num, (0 to stop)"))
    if num==0:
        break
````

# List comprehension

````python
%%time

# unoptimized function
square = []
for i in range(10):
    if i%2==0:
        square.append(i*i)
print(square)
````

````python
%%time

square = [i*i for i in range(10) if i%2==0 ]
print(square)
````

# QST 5

```py
name = ["Rama", "sita", "lakshmana", "john"]
```

Get me those name that has only 'a' in it

````python
names = ["Rama", "sita", "lakshmana", "john"]
word = [i for i in names if "a" in i]
print(word)
````

````python
# 5.b > print the last character using list comprehension

lastChar = [ i[-1] for i in names ]
print(lastChar)
````

````python
# 5.c > print name that have more than 4 characters using list comprehension

lenGreaterThanFour = [i for i in names if len(i)>4]
print(lenGreaterThanFour)
````

````python
# 5.d > print word and its length using list comprehension

wordCount = [{i,len(i)} for i in names]
print(wordCount)
````

````python
# 5.e > convert each word to upper case using list comprehension

upper = [str.upper(i) for i in names]
print(upper)
````

````python
# 5.f > cover name only if it has length > 4 using list comprehension

upperGrtThan4 = [str.upper(i) for i in names if len(i)>4]
print(upperGrtThan4)
````

````python
# 5.g > give common elements in both list using list comprehension

l = [1,2,3,4]
l1=[3,4,5,6]

common = [i for i in l1 if i in l]
print(common)
````
