## QST 1 - Inputs
Using input function, create a list base on the input provided by user and get inputs for the list too

```python
n = int(input("enter list length: "))
arr = []
for i in range(n):
    val=int(input("enter input for index " + str(i) + " : "))
    arr.append(val)

print(arr)
```

using while loop

```python
n = int(input("enter list length: "))
arr = []
while n>0:
    val = int(input("enter input for index " + str(n) + " : "))
    arr.append(val)
    n=n-1
print(arr)
```

using list comprehension

```python
n,arr = int(input("enter list length: ")),[]
[arr.append(int(input("enter input for index " + str(i) + " : "))) for i in range(n)]
print(arr)
```

**Efficient way**:
Often in competitive programming or data processing, it's more efficient to ask the user for all numbers at once (separated by spaces) rather than hitting the input() syscall n times.

```python
# Input: 10 20 30 40
arr = list(map(int, input("Enter all numbers separated by space: ").split()))
```

## QST 2 - While
Sum of 10 number

```python
n = 10
sum = 0
while n >0:
    sum+=n
    n=n-1
print("sum: ",sum)
```

Usage of while

```python
num = int(input("Enter the num, (0 to stop)"))
while num != 0:
    print("Entered number is: ", num)
    num=int(input("Enter the num, (0 to stop)"))
```

Usage of while with break

```python
num = int(input("Enter the num, (0 to stop)"))
while True:
    print("Entered number is: ", num)
    num=int(input("Enter the num, (0 to stop)"))
    if num==0:
        break
```
