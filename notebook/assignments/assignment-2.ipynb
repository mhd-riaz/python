Company has 5 departments and 6 employee each.
1. Generate salary matrix.
2. Find:
   1. Department-wise average salary
   2. Employee-wise maximum salary
3. Increase all salaries by 10%
4. Find total salary expenditure.

Thoughts:
1. 5 department - 5 columns
2. 6 employee - 6 rows
3. for each employee use np.random.randint for generating salary in given range

```python
# import statements
import numpy as np
np.random.seed()
```

```python
# generate salary for each employee
# i.e. start, stop(excluding), (6 rows, 5 col)
salary = np.random.randint(30000, 1200000, (6,5))
print(salary)
```

```python
# Find Avg per department
avg = np.mean(salary, axis=0)
print("Avg: ",avg)
```

```python
# Find employee wise max salary
max = np.max(salary, axis=1)
print("Max: ",max)
```

```python
# increase all salaries by 10%
salary=salary+(salary*(10/100))
print("new salary: ", salary)
```

```python
# total salary
totalByDepartment = np.sum(salary,axis=1)
print(totalByDepartment)

# overallTotal
print("overall salary expenditure: ", np.sum(totalByDepartment))
```
