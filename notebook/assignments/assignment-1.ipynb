A class has 8 students and 4 subjects.
1. Generate marks randmly between 40 and 100.
2. Find:
   - Average marks per student
   - Subject-wise highest marks
   - Student who scored highest total
3. Replace all marks below 50 with 0.

Thoughts:
1. 8 students - 8 rows
2. 4 subjects - 4 columns
3. for each subject use np.random.randint for generating marks in given range

```python
# import statements
import numpy as np
np.random.seed()
```

```python
# generate marks for each student
# i.e. start, stop(excluding), (8 rows, 4 col)
marks = np.random.randint(40, 101, (8,4))
print(marks)
```

```python
# Find Avg per student
avg = np.mean(marks, axis=1)
print("Avg: ",avg)
```

```python
# Find Highest marks per subject
max = np.max(marks, axis=0)
print("Max: ",max)
```

```python
# Find total per student
total = np.sum([marks], axis=0)
print("Total: ", total)

# Find student who scored highest mark
student_index = np.argmax(total)
print("Student who scored highest mark is at index: ", student_index)
print("Mark of the student: ", marks[student_index])
```

```python
# replace marks < 50 with 0
marks[marks<50] = 0
print("New marks: ", marks)
```
