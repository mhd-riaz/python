Mathematical Example<br>
If you have scores: [80, 90, 90, 100]<br>
Assign Raw Positions:
- 80 is at position
- 1.The two 90s are at positions 2 and 3
- 100 is at position 4.
  
Compute Rank (using average):
- The 90s occupy spots 2 and 3.
- $\text{Average Rank} = \frac{2 + 3}{2} = 2.5
- $Final Ranks: [1.0, 2.5, 2.5, 4.0]

<!-- Rank parameters:

- method: [‘average’, ‘min’, ‘max’, ‘first’, ‘dense’]
- ascending: bool -->

Display only students from CSE department.

Add a new column "Result"

Marks ≥ 85 "Pass"

Else → "Needs Improvement"

Sort dataset in descending order of Marks Find minimum marks and display student name Count how many students in each department Display only Name and Marks columns

```python
# 1. Load the dataset
df = pd.read_csv("./notebook/students.csv")
df.head()
```

```python
# Column names:
print("Column names: ", df.columns)
```

```python
# avg mark of all student:
avgMarks = df['Marks'].mean()
print(avgMarks)
```

```python
# highest marks
print(df.loc[df['Marks'].idxmax()])
```

```python
# filter CSE student only
print(df.loc[df['Dept']=="CSE"])
```

```python
# add new column called result
df["Result"] = df["Marks"].apply(lambda x: "PASS" if x >=85 else "Needs Improvement")

df.head()
```

```python
# Sort dataset in descending order of Marks Find minimum marks and display student name

# Sort in descending order
df.sort_values("Marks",ascending=False, inplace=True)
df.head()

# Get min mark
df.tail(1) # if you did inplace sort
df['Name'][df['Marks'].idxmin()]
```

```python
#  Count how many students in each department
df['Dept'].value_counts()
# df.groupby("Dept").count()
```
