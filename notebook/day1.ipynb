### Reversing a word

```python
# 1 way - using for loop
input = "we love python"
word = input.split(" ")
for w in range(len(word)):
    reverse = word[w][::-1]
    print(reverse, end=" ")
```

```python
# 2nd way - using list comprehension
reverse = [i[::-1] for i in (input).split(" ")]
" ".join(reverse)

# is better in terms of complexity than for loop, as we use compressed version of the same
```

### find the common elements in both list

```python
l1 = [10, 22, 34, 55]
l2 = [5, 10, 22, 77]
common=[]
for i in l1:
    if i in l2:
        common.append(i)
common
```

```python
# approach 2
print([i for i in l1 if i in l2 ])
```
