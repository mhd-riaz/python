### Reversing a word

```python
# 1 way - using for loop
input = "we love python"
word = input.split(" ")
for w in range(len(word)):
    reverse = word[w][::-1]
    print(reverse, end=" ")
```

```python
# 2nd way - using list comprehension
reverse = [i[::-1] for i in (input).split(" ")]
" ".join(reverse)

# is better in terms of complexity than for loop, as we use compressed version of the same
```

### find the common elements in both list

```python
l1 = [10, 22, 34, 55]
l2 = [5, 10, 22, 77]
common=[]
for i in l1:
    if i in l2:
        common.append(i)
common
```

find total runs scored by kohli, where<br> runs=["kohli", 100, 20, 0, 60]

```python
# approach 1; using index
runs=["kohli", 100, 20, 0, 60]
score = 0
for i in range(0,len(runs)):
    if type(runs[i]) == int:
        score+=int(runs[i])
score
```

```python
# approach 2; using values
score = 0
for i in runs:
    if type(i) == int:
        score+=i
score
```

```python
# if you know when the key is
# approach 3
score = 0
for i in runs[1:]:
    score +=i
score
```

```python
# if you know when the key is
# approach 4
score = 0
for i in runs[-(len(runs)-1):]: # or `for i in runs[-4:]:`
    score +=i
score
```

### For each word create a count and store in dict<br>

`input: "finally we are coming to the of the class"`

```python
s = "finally we are coming to the of the class"

# approach 1 - without methods
d = {}
for i in s.split(" "):
    if i not in d:
        d[i] = 0
    d[i] = d[i]+1
d
```

```python
# approach 2 - using methods
counts = {}
print(s)
for i in s.split(" "):
    counts[i] = i.count(i)
print(counts)
```
